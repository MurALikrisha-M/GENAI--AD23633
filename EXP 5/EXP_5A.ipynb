### 1. Install and Import Libraries

```python
import warnings
warnings.filterwarnings('ignore')

from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
```

### 2. Create Sample Text Document


```python
with open("sample.txt", "w", encoding="utf-8") as f:
    f.write("Artificial Intelligence (AI) is a field of computer science that focuses on creating machines capable of performing tasks that typically require human intelligence.\n\nMachine learning is a subset of AI that allows systems to learn from data and improve over time without being explicitly programmed.\n\nNatural Language Processing (NLP) enables computers to understand, interpret, and generate human language.\n\nDeep learning is a specialized area of machine learning that uses neural networks with many layers to analyze complex patterns in data.\n\nAI is widely used in applications such as chatbots, recommendation systems, image recognition, and autonomous vehicles.\n")
```

### 3. Load the Document


```python
loader = TextLoader("sample.txt", encoding="utf-8")
documents = loader.load()
```

### 4. Split the Document


```python
splitter = CharacterTextSplitter(
    separator="\n",
    chunk_size=500,
    chunk_overlap=50
)
split_docs = splitter.split_documents(documents)
```

### 5. Create Embeddings

```python
embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)
```

### 6. Create FAISS Vector Store



```python
vectorstore = FAISS.from_documents(
    documents=split_docs,
    embedding=embeddings
)
```

### 7. Perform Similarity Search

```python
query = "What is this document about?"
results = vectorstore.similarity_search(query, k=3)
```

### 8. Output Search Results

```python
for i, doc in enumerate(results, start=1):
    print(f"\nResult {i}")
    print("-" * 50)
    print(doc.page_content)
    print("Metadata:", doc.metadata)
```
